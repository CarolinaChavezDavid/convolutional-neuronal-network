# Visualizar ejemplos del dataset

Este notebook selecciona 6 imagenes de clases distintas dentro de esta carpeta, las muestra en una grilla de 3 imagenes por fila y guarda la figura resultante en `ejemplos_dataset.png`.

Estructura esperada:

```text
dataset/
  clase_1/
    imagen_1.jpg
  clase_2/
    imagen_2.jpg
```


````python
from pathlib import Path
import random

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
````

````python
DATASET_DIR = Path.cwd()
OUTPUT_PATH = DATASET_DIR / "ejemplos_dataset.png"

IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".webp", ".tif", ".tiff"}
N_EXAMPLES = 6
N_COLS = 3
RANDOM_SEED = 42
````

````python
def image_files_for_class(class_dir: Path):
    return sorted(
        path
        for path in class_dir.rglob("*")
        if path.is_file() and path.suffix.lower() in IMAGE_EXTENSIONS
    )


classes_with_images = []
for class_dir in sorted(path for path in DATASET_DIR.iterdir() if path.is_dir()):
    images = image_files_for_class(class_dir)
    if images:
        classes_with_images.append((class_dir.name, images))

print(f"Clases con imagenes encontradas: {len(classes_with_images)}")
for class_name, images in classes_with_images:
    print(f"- {class_name}: {len(images)} imagen(es)")
````

````python
if len(classes_with_images) < N_EXAMPLES:
    raise ValueError(
        f"Se necesitan al menos {N_EXAMPLES} clases con imagenes, "
        f"pero se encontraron {len(classes_with_images)}."
    )

rng = random.Random(RANDOM_SEED)
selected_classes = rng.sample(classes_with_images, N_EXAMPLES)
selected_examples = [
    (class_name, rng.choice(images))
    for class_name, images in selected_classes
]

selected_examples
````

````python
n_rows = (N_EXAMPLES + N_COLS - 1) // N_COLS
fig, axes = plt.subplots(n_rows, N_COLS, figsize=(12, 8))
axes = axes.flatten()

for ax, (class_name, image_path) in zip(axes, selected_examples):
    image = mpimg.imread(image_path)
    ax.imshow(image)
    ax.set_xlabel(class_name, fontsize=13, labelpad=10)
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)

for ax in axes[len(selected_examples):]:
    ax.axis("off")

fig.tight_layout()
fig.savefig(OUTPUT_PATH, dpi=200, bbox_inches="tight")
plt.show()

print(f"Imagen guardada en: {OUTPUT_PATH}")
````
